# Convert

```python
def modify(v):
    if v == '13Threes':
        return 13.3
    elif v == '12Twos':
        return 12.2
    elif v == '14Threes':
        return 14.3
    elif v == '13Ones':
        return 13.1
    elif v == '14Ones':
        return 14.1
    elif v == '14Twos':
        return 14.2
    elif v == '12Ones':
        return 12.1
    elif v == '11Ones':
        return 11.1
    elif v == '13Twos':
        return 13.2
    elif v == '10Ones':
        return 10.1
    elif v == '11Twos':
        return 11.2
    elif v == '12Fours':
        return 12.4
    elif v == '12Threes':
        return 12.3
    
    elif v == '13Fives':
        return 13.5
    elif v == '13Fours':
        return 13.4
    elif v == '14Sixes':
        return 14.6
    elif v == '15Sixes':
        return 15.6
    elif v == '16Ones':
        return 16.1
    elif v == '14Fours':
        return 14.4
    elif v == '16Threes':
        return 16.3
    elif v == '16Twos':
        return 16.2
    elif v == '11Threes':
        return 11.3
    elif v == '14Fives':
        return 14.5
    
    elif v == '15Fives':
        return 15.5
    elif v == '15Fours':
        return 15.4
    elif v == '15Ones':
        return 15.1
    elif v == '15Threes':
        return 15.3
    elif v == '15Twos':
        return 15.2
    else:
        return(0)
```

```python
def convert(a):
    if a == 'OH': return 1
    elif a == 'ALL': return 2
    elif a == 'Hitter': return 3
    elif a == 'Libero': return 4
    elif a == 'Middle': return 5
    elif a == 'Opposite': return 6
    elif a == 'Setter': return 7
    elif a == 'Utility': return 0
```

```python
def convert_month(a):
    if a == 'Jan': return 1
    elif a == 'Feb': return 2
    elif a == 'Mar': return 3
    elif a == 'Apr': return 4
    elif a == 'May': return 5
    elif a == 'Jun': return 6
    elif a == 'Jul': return 7
    elif a == 'Aug': return 8
    elif a == 'Sep': return 9
    elif a == 'Oct': return 10
    elif a == 'Nov': return 11
    elif a == 'Dec': return 12
```

```python
def convert_age(a):
    return int(a[:2])
```

```python
df['Team'] = df['Team'].apply(modify)
```

```python
pos_list = []
for i in df['Postition1']:
    if 'OH' in i:
        pos_list.append('OH')
    else:
        pos_list.append(i)
df['Postition1'] = pos_list
```

```python
df['Age_int'] = df['Age'].apply(convert_age)
```

```python
df['Mon_int'] = df['Month'].apply(convert_month)
```

# Neural Network

```python
x = df[['Mon_int', 'Age_int', 'Hit', 'Serve', 'Set', 'Throw', 'Pass', 'Total', 'Jump', 'AppVertical', 'Height', 'Reach', 'Speed', 'Team']].values # Could add month & team
y = df['Postition1'].apply(convert)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python

y_train = pd.get_dummies(y_train).values

y_test = pd.get_dummies(y_test).values

X_train.shape, y_train.shape
```

```python
import os
os.environ['KERAS_BACKEND']='tensorflow'
import keras as ks
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(14, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(14, activation='relu'))
model.add(Dense(8, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=10)
```

```python
_, accuracy = model.evaluate(X_test, y_test)
```

# Dashboard Function Based on the Above

```python
def dashboard(stat, epo, batch):
    x = df[['Mon_int', 'Age_int', 'Hit', 'Serve', 'Set', 'Throw', 'Pass', 'Total', 'Jump', 'AppVertical', 'Height', 'Reach', 'Speed', 'Team']].values # Could add month & team
    y = df['Postition1'].apply(convert)

    #Scale x & stat
    scaler = MinMaxScaler()
    scaler.fit(x)
    x = scaler.transform(x)
    stat = scaler.transform(stat)


    #Create dummies
    y_train = pd.get_dummies(y).values
    
    #Model
    model = Sequential()
    model.add(Dense(14, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(12, activation='relu'))
    model.add(Dense(8, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(x, y_train, epochs=epo, batch_size=batch)
    accuracy = model.predict_classes(stat)
    
    #Output position
    output = []
    for a in accuracy:
        if a == 1: output.append('OH')
        elif a == 2: output.append('ALL')
        elif a == 3: output.append('Hitter')
        elif a == 4: output.append('Libero')
        elif a == 5: output.append('Middle')
        elif a == 6: output.append('Opposite')
        elif a == 7: output.append('Setter')
        elif a == 0: output.append('Utility')
    print('-----------------------------------------')      
    print("The player(s) predicted positions are")
    return output
```
